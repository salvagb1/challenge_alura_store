### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/salvagb1/challenge_alura_store/main/data_base/tienda_1.csv"
url2 = "https://raw.githubusercontent.com/salvagb1/challenge_alura_store/main/data_base/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/salvagb1/challenge_alura_store/main/data_base/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/salvagb1/challenge_alura_store/main/data_base/tienda_4.csv"

# Eliminar refs/heads/ del link que te genera cuando le das clic en raw del csv

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Ingreso total por cada tienda



```python
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

print(f"Ingreso total tienda 1: ${ingreso_tienda_1}")
print(f"Ingreso total tienda 2: ${ingreso_tienda_2}")
print(f"Ingreso total tienda 3: ${ingreso_tienda_3}")
print(f"Ingreso total tienda 4: ${ingreso_tienda_4}")
```

```python
# Ingresos por tienda
ingresos = [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Gráfico de barras para el ingreso total
plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color='blue')

# Añadir etiquetas y título
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso ($)')
plt.tight_layout()
plt.show()
```

```python
# Datos de ingresos por tienda (reemplaza estos valores con los datos reales)
ingresos = [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Gráfico de barras horizontales
plt.figure(figsize=(10, 5))
bars = plt.barh(tiendas, ingresos, color='purple')

# Añadir etiquetas y título
plt.title('Ingreso Total por Tienda')
plt.xlabel('Ingreso ($)')
plt.ylabel('Tienda')

# Agregar los valores de ingreso sobre las barras
for bar in bars:
    plt.text(bar.get_width() + 1000, bar.get_y() + bar.get_height()/2, f'${bar.get_width():,.0f}',
             va='center', ha='left', color='black')

plt.tight_layout()
plt.show()
```

# 3. Valoración media por tienda


```python
# Paso 1: Sumar los valores
suma_calificaciones = tienda['Calificación'].sum()

# Paso 2: Contar cuántos valores hay
cantidad_calificaciones = tienda['Calificación'].count()

# Paso 3: Dividir para obtener el promedio
promedio_manual = suma_calificaciones / cantidad_calificaciones

print(f"Promedio (calculado manualmente) de tienda 1: {promedio_manual:.2f}")
print(' ')

#Con .mean() toma todos los valores numéricos de una columna (en este caso, la columna "Calificación") y calcula el promedio
promedio_tienda_1 = tienda['Calificación'].mean()
promedio_tienda_2 = tienda2['Calificación'].mean()
promedio_tienda_3 = tienda3['Calificación'].mean()
promedio_tienda_4 = tienda4['Calificación'].mean()

print(f"Promedio de calificación de Tienda 1: {promedio_tienda_1:.2f}")
print(f"Promedio de calificación de Tienda 2: {promedio_tienda_2:.2f}")
print(f"Promedio de calificación de Tienda 3: {promedio_tienda_3:.2f}")
print(f"Promedio de calificación de Tienda 4: {promedio_tienda_4:.2f}")
```

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios = [promedio_tienda_1, promedio_tienda_2, promedio_tienda_3, promedio_tienda_4]
```

```python
# Gráfico
plt.bar(tiendas, promedios, color='teal')
plt.ylim(0, 5)
plt.title('Promedio de calificación por tienda')
plt.ylabel('Calificación promedio')

# Agregar etiquetas con los valores encima de cada barra
for i, valor in enumerate(promedios):
    plt.text(i, valor + 0.05, f"{valor:.2f}", ha='center')

plt.tight_layout()
plt.show()
```

```python
# Gráfico de barras horizontales
plt.figure(figsize=(8, 5))
plt.barh(tiendas, promedios, color='mediumseagreen')
plt.title('Calificación promedio por tienda (horizontal)')
plt.xlabel('Calificación')

# Mostrar los valores en las barras
for i, valor in enumerate(promedios):
    plt.text(valor + 0.05, i, f"{valor:.2f}", va='center')

plt.xlim(0, 5)
plt.tight_layout()
plt.show()
```

```python
# Total de ventas por zona (sin importar tienda)
zona_ventas = df_total.groupby(['lat', 'lon']).size().reset_index(name='ventas')

# Tienda con más ventas por zona
zona_ventas_tienda = df_total.groupby(['lat', 'lon', 'tienda']).size().reset_index(name='ventas_tienda')
zona_top_tienda = zona_ventas_tienda.sort_values('ventas_tienda', ascending=False).drop_duplicates(subset=['lat', 'lon'])

# Unir ambos para que tengas en una sola tabla las ventas totales y la tienda dominante
zona_ventas_completa = pd.merge(zona_ventas, zona_top_tienda, on=['lat', 'lon'])
zona_ventas_completa = zona_ventas_completa[['lat', 'lon', 'ventas', 'tienda', 'ventas_tienda']]

# Ordenar por zonas más activas
zona_ventas_completa.sort_values(by='ventas', ascending=False).head()
```
